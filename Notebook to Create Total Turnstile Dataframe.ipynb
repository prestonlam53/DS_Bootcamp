```python
import pandas as pd
```

```python
#read in data
df_census = pd.read_csv('acs2017_census_tract_data.csv')
```

```python
#filter for only data from NYC
df_census = df_census[df_census['State'] == 'New York']
df_census = df_census[df_census['County'].isin(['New York County', 'Kings County', 'Bronx County', 'Richmond County', 'Queens County'])]
```

```python
#import and combine all turnstile files
import glob
import pandas as pd
extension = 'txt'
all_filenames = [i for i in glob.glob('turnstile*.{}'.format(extension))]
print(all_filenames)


#combine all files in the list
combined_turnstile_csv = pd.concat([pd.read_csv(f) for f in all_filenames ])
```

```python
combined_turnstile_csv
```

```python
#Get subway tract IDs and merge them with overall csv
subway_tracts = pd.read_csv('nyc_subway_final.csv')
subway_tracts['Station Name'] = subway_tracts['Station Name'].str.lower()
subway_tracts.set_index('Station Name', inplace = True)

combined_turnstile_csv['STATION'] = combined_turnstile_csv['STATION'].str.lower()
combined_turnstile_csv.rename(columns = {'STATION': 'Station Name'}, inplace = True)
```

```python
combined_turnstile_csv = pd.merge(combined_turnstile_csv,subway_tracts, on = 'Station Name', how = 'outer')

combined_turnstile_csv
```
